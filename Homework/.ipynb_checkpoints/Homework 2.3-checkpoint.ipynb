### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
dim_x = 100
dim_y = 100

U = np.zeros((dim_x, dim_y))

xs = np.linspace(0, 1, dim_x)
ys = np.linspace(0, 1, dim_y)

U[:, 0] = np.sin(2*np.pi*xs)
U[:, -1] = np.sin(2*np.pi*xs)
U[0, :] = np.sin(3*np.pi*ys)
U[-1, :] = np.sin(np.pi*ys)
```

```python
error = 1
while error > 0.001:
    next_U = U.copy()
    for i in range(1, dim_x-1):
        for j in range(1, dim_y-1):
            next_U[i, j] = (1./4) *(U[i+1, j] + U[i-1, j] + U[i, j+1] + U[i, j-1])
    delta_U = next_U - U
    error = np.linalg.norm(delta_U)
    U = next_U.copy()
```

```python
plt.imshow(U, cmap='hot')
```
